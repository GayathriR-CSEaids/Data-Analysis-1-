### **LINE** **PLOT**

```python
import pandas as pd
import plotly.graph_objects as go


# Calculate total cyclonic activity
df['Total'] = df['Cyclonic Disturbances - TOTAL'] + df['Cyclones - TOTAL'] + df['Severe Cyclones - TOTAL']

# Create figure
fig = go.Figure()

# Add traces for each data series
fig.add_trace(go.Scatter(x=df['Year'], y=df['Cyclonic Disturbances - TOTAL'], mode='lines+markers', name='Cyclonic Disturbances - TOTAL'))
fig.add_trace(go.Scatter(x=df['Year'], y=df['Cyclones - TOTAL'], mode='lines+markers', name='Cyclones - TOTAL'))
fig.add_trace(go.Scatter(x=df['Year'], y=df['Severe Cyclones - TOTAL'], mode='lines+markers', name='Severe Cyclones - TOTAL'))
fig.add_trace(go.Scatter(x=df['Year'], y=df['Total'], mode='lines+markers', name='Total Cyclonic Activity', line=dict(dash='dash')))

# Update layout with range slider
fig.update_layout(
    title='Total Cyclonic Activity (1891-1900)',
    xaxis_title='Year',
    yaxis_title='Count',
    xaxis=dict(
        rangeslider=dict(visible=True),
        type='linear'  # Ensure the x-axis is treated as linear
    ),
    yaxis=dict(range=[0, df['Total'].max() + 5]),  # Adjust y-axis range for better visibility
)

# Show the plot
fig.show()
```

### **TREND** **LINE**

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go


# Create a figure
fig = go.Figure()

# Add traces for each data series
for column in ['Cyclonic Disturbances - TOTAL', 'Cyclones - TOTAL', 'Severe Cyclones - TOTAL']:
    # Add line trace
    fig.add_trace(go.Scatter(
        x=df['Year'],
        y=df[column],
        mode='lines+markers',
        name=column
    ))

    # Calculate the trend line
    z = np.polyfit(df['Year'], df[column], 1)  # Linear fit
    p = np.poly1d(z)

    # Add trend line
    fig.add_trace(go.Scatter(
        x=df['Year'],
        y=p(df['Year']),
        mode='lines',
        name=f'Trend Line - {column}',
        line=dict(dash='dash')  # Dashed line for the trend
    ))

# Update layout
fig.update_layout(
    title='Total Cyclonic Activity with Trend Lines (1891-1900)',
    xaxis_title='Year',
    yaxis_title='Count',
    xaxis=dict(range=[1891, 1900]),
    yaxis=dict(range=[0, df[['Cyclonic Disturbances - TOTAL', 'Cyclones - TOTAL', 'Severe Cyclones - TOTAL']].max().max() + 5]),
)

# Show the plot
fig.show()
```

### **CORRELATION MATRIX**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'df' is your DataFrame
# Compute the correlation matrix
correlation_matrix = df.corr()

# Set the size of the plot
plt.figure(figsize=(10, 8))

# Generate a heatmap for the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

# Set the title
plt.title('Correlation Matrix Heatmap', size=16)

# Display the heatmap
plt.show()
```

Cyclonic Disturbances - BOB and Severe Cyclones - BOB have a strong positive correlation of 0.90, indicating that higher cyclonic disturbances in the Bay of Bengal tend to be associated with more severe cyclones.

Cyclonic Disturbances - BOB and Cyclones - TOTAL have a correlation of 0.62, suggesting that cyclonic disturbances in the Bay of Bengal contribute significantly to the total number of cyclones.

Year is negatively correlated with most variables, but it's particularly negatively correlated with Cyclonic Disturbances - TOTAL and Severe Cyclones - AS, meaning that over time, these values decreased in the sample period.

### **STACKED** **PLOT**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is already defined with your dataset

# Define the regions and categories for comparison
regions = ['BOB', 'AS', 'LAND']
categories = ['Cyclonic Disturbances', 'Cyclones', 'Severe Cyclones']

# Create a dictionary to store the data for the stacked bar chart
stacked_data = {
    'BOB': [df['Cyclonic Disturbances - BOB'].sum(), df['Cyclones - BOB'].sum(), df['Severe Cyclones - BOB'].sum()],
    'AS': [df['Cyclonic Disturbances - AS'].sum(), df['Cyclones - AS'].sum(), df['Severe Cyclones - AS'].sum()],
    'LAND': [df['Cyclonic Disturbances - LAND'].sum(), df['Cyclones - LAND'].sum(), df['Severe Cyclones - LAND'].sum()]
}

# Convert the dictionary into a DataFrame
stacked_df = pd.DataFrame(stacked_data, index=categories)

# Plot the stacked bar chart
stacked_df.T.plot(kind='bar', stacked=True, figsize=(10, 7), color=['#40E0D0', 'magenta', '#007BFF'])

# Set the labels and title
plt.ylabel('Number of Storms')
plt.title('Composition of Storms by Region')
plt.xticks(rotation=0)
plt.legend(title='Storm Category')

# Show the plot
plt.show()
```

```python
# Data for the grouped bar chart
categories = ['Cyclonic Disturbances', 'Cyclones', 'Severe Cyclones']
regions = ['BOB', 'AS', 'LAND']
width = 0.25  # Width of the bars

# Values for each category in each region
bob_values = [df['Cyclonic Disturbances - BOB'].sum(), df['Cyclones - BOB'].sum(), df['Severe Cyclones - BOB'].sum()]
as_values = [df['Cyclonic Disturbances - AS'].sum(), df['Cyclones - AS'].sum(), df['Severe Cyclones - AS'].sum()]
land_values = [df['Cyclonic Disturbances - LAND'].sum(), df['Cyclones - LAND'].sum(), df['Severe Cyclones - LAND'].sum()]

# Positions for the bars
x = np.arange(len(categories))

# Plotting grouped bars for each intensity level by region
fig, ax = plt.subplots(figsize=(10, 7))

# Plot data for each region
ax.bar(x - width, bob_values, width, label='BOB')
ax.bar(x, as_values, width, label='AS')
ax.bar(x + width, land_values, width, label='LAND')

# Set labels and title
ax.set_xlabel('Intensity Level')
ax.set_ylabel('Number of Storms')
ax.set_title('Intensity Levels by Region')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

# Show the plot
plt.show()
```

### **BOX** **PLOT**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assuming 'df' is your DataFrame with relevant data

# Reshape the data using melt
plot_df = pd.melt(df,
                  id_vars=['Year'], # Assuming 'Year' is a column in your DataFrame
                  value_vars=['Cyclonic Disturbances - BOB', 'Cyclonic Disturbances - AS', 'Cyclonic Disturbances - LAND',
                             'Cyclones - BOB', 'Cyclones - AS', 'Cyclones - LAND',
                             'Severe Cyclones - BOB', 'Severe Cyclones - AS', 'Severe Cyclones - LAND'],
                  var_name='Category_Region',
                  value_name='Values')

# Extract Category and Region from 'Category_Region'
plot_df[['Category', 'Region']] = plot_df['Category_Region'].str.split(' - ', expand=True)

# Set the figure size
plt.figure(figsize=(10, 7))

# Create a box plot
sns.boxplot(x='Category', y='Values', hue='Region', data=plot_df)

# Set labels and title
plt.xlabel('Intensity Level')
plt.ylabel('Number of Storms')
plt.title('Box Plot of Storm Intensities by Region')

# Display the plot
plt.show()
```

### **SUN BURST**

```python
import plotly.express as px

# Prepare data for sunburst chart
sunburst_data = {
    'Region': ['Bay of Bengal'] * 3 + ['Arabian Sea'] * 3 + ['Land'] * 3,
    'Storm Type': ['Cyclonic Disturbances', 'Cyclones', 'Severe Cyclones'] * 3,
    'Counts': bob_values + as_values + land_values
}

sunburst_df = pd.DataFrame(sunburst_data)

# Create a sunburst chart
fig = px.sunburst(sunburst_df, path=['Region', 'Storm Type'], values='Counts',
                  title='Sunburst Chart of Storm Types by Region')
fig.show()
```

### **TREE** **MAP**

```python
import plotly.graph_objects as go


# Combine the values and labels
sizes = bob_values + as_values + land_values
labels = ['Cyclonic Disturbances - BOB', 'Cyclones - BOB', 'Severe Cyclones - BOB',
          'Cyclonic Disturbances - AS', 'Cyclones - AS', 'Severe Cyclones - AS',
          'Cyclonic Disturbances - LAND', 'Cyclones - LAND', 'Severe Cyclones - LAND']
parents = ['', '', '', '', '', '', '', '', '']  # Set empty strings as all are at the root level

# Create an interactive treemap using Plotly
fig = go.Figure(go.Treemap(
    labels=labels,
    values=sizes,
    parents=parents,
    marker=dict(colors=sizes, colorscale='Plasma'),  # Use 'Viridis' for a visually appealing color scale
    textinfo="label+value",  # Display both labels and values
    hoverinfo="label+value+percent parent",  # Interactive hover info
))

# Add a title and display the treemap
fig.update_layout(
    title='Interactive Treemap of Storm Counts by Type and Region',
    title_font_size=20,
    width=800,  # Adjust width
    height=500,  # Adjust height
    margin=dict(t=50, l=25, r=25, b=25)  # Adjust margins for a cleaner look
)

fig.show()
```
